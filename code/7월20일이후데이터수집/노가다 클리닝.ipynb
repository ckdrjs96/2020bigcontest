```python

df['G_DAY_DS']=data['날짜']
df['T_ID']=data['팀명']
df['VS_T_ID']=data['상대팀']
df['HEADER_NO']=0
df['TB_SC']=data['초말']
df['PA']=data['타석']
df['AB']=data['타수']
df['RBI']=data['타점']
df['RUN']=data['득점']
df['HIT']=data['안타']
df['H2']=data['2루타']
df['H3']=data['3루타']
df['HR']=data['홈런']
df['SB']=data['도루']
df['CS']=data['도실']
df['SH']=data['희타']
df['SF']=data['희플']
df['BB']=data['볼넷']
df['IB']=data['고4']
df['HP']=data['HBP']
df['KK']=data['삼진']
df['GD']=data['병살']
df['ERR']=data['병살']
df['LOB']=data['잔루']
df['P_HRA_RT']=data['득점권타율']
df['P_AB_CN']=data['득점권타수']
df['P_HIT_CN']=data['득점권안타']
df['G_ID']=df.reset_index()['index']

```
