## 1. AVG

```python
hiter16['AVG']= hiter16['HIT']/hiter16['AB']
```

## 2. OPS

```python
#2. OPS = 출루율 + 장타율
# 출루율
hiter16['OBP'] =(hiter16['HIT']+hiter16['BB']+hiter16['HP'])/(hiter16['AB']+hiter16['BB']+hiter16['HP']+hiter16['SF'])
```

```python
# 장타율
hiter16['SLG'] = (hiter16['HIT']+hiter16['H2']+2*hiter16['H3']+3*hiter16['HR'])/hiter16['AB']
```

```python
# OPS 
hiter16['OPS']=hiter16['AVG']+hiter16['OBP']
```

## 3. wOBA 

```python
#3. wOBA= [{(볼넷-고의4구)x0.7}+(몸에 맞는 볼 x 0.73)+(1루타x0.89)+
#(2루타x1.27)+(3루타 x 1.61)+(홈런x2.07)+(도루x0.25)+(도실x0.5)](타석-고의4구)
hiter16['H1']=hiter16['HIT']-(hiter16['H2']+hiter16['H3']+hiter16['HR'])
hiter16['wOBA']= (((hiter16['BB']-hiter16['IB'])*0.7) + (hiter16['HP']*0.73)+ (hiter16['H1']*0.89)+(hiter16['H2']*1.27)+(hiter16['H3']*1.61)+(hiter16['HR']*2.07)+(hiter16['SB']*0.25)+(hiter16['CS']*0.5))/(hiter16['AB']-hiter16['IB'])
```

## 4. wRC 

```python
#4. wRC = {((wOBA - 리그평균wOBA) / wOBA Scale) + (리그득점/리그타석)} * 타석
wOBA_scale = 1.1 # 얘는 임의로 설정하는건가?
sum_df = hiter16.sum()
rpa = sum_df['RUN'] / sum_df['AB']
hiter16['wRC'] =  ((hiter16['wOBA']-np.mean(hiter16['wOBA']))/wOBA_scale)+rpa*hiter16['AB']
```

## Score 정의

```python
# 스코어 계산해내는 함수
from sklearn.metrics import make_scorer
def mae(prediction, correct):
    prediction = np.array(prediction)
    correct = np.array(correct)
    
    difference = correct - prediction
    abs_val = abs(difference)
    
    score = abs_val.mean()
    
    return score

mae_scorer = make_scorer(mae)
mae_scorer
```

## 그래프 그리는 함수 정의

```python
# 그래프 그리는 함수
import matplotlib.pyplot as plt
def plot_train_test_pred_graph(trainset, testset, pred):
    plt.figure(figsize=(15,3))
    plt.plot(trainset.RUN, label='train')
    plt.plot(testset.RUN, label='test')
    plt.plot(testset.index, pred, label='prediction')
    plt.legend()
    plt.show()
```

```python
team_list = ['WO','OB','NC','SK','LG','KT','LT','HT','SS','HH']
```

# Simple Exponential Smoothing

```python
hiter_df
```

```python
def simpleExpSmoothing(df, team):
    team = df[df['T_ID']==team]
    df_set = team
    size = int(len(df_set) * 0.85)
    train = df_set[:size]
    test = df_set[size:]
    
    ses_model = SimpleExpSmoothing(train.RUN)
    ses_result = ses_model.fit()
    ses_pred = ses_result.forecast(len(test))
    print("store_id 0 mean value : ", df_set.RUN.mean())
    print("MAE Score of test :", mae(test.RUN, ses_pred))
    plot_train_test_pred_graph(train, test, ses_pred)

    return ses_pred
```

```python
pred_list=[]
for team in team_list:
    print(team)
    pred=simpleExpSmoothing(hiter_df, team)
    pred_list.append(pred)
```

## Holt's Exponential Smoothing

```python
def HoltsExpSmoothing(df,team):
    team = df[df['T_ID']==team]
    df_set = team
    size = int(len(df_set) * 0.85)
    train = df_set[:size]
    test = df_set[size:]
    holt_model = Holt(np.array(train.RUN))
    holt_result = holt_model.fit()
    holt_pred = holt_result.forecast(len(test))
    
    print("Mean value of store_id 0 : ", df_set.RUN.mean())
    print("MAE Score of test :", mae(test.RUN, holt_pred))

    plot_train_test_pred_graph(train, test, holt_pred)
    
    return holt_pred
```

```python
pred_list=[]
for team in team_list:
    print(team)
    pred=HoltsExpSmoothing(hiter_df, team)
    pred_list.append(pred)
```

## Holt-Winter's Exponential Smoothing

```python
def HoltsWintersExpSmoothing(df,team):
    team = df[df['T_ID']==team]
    df_set = team

    size = int(len(df_set) * 0.85)
    train = df_set[:size]
    test = df_set[size:]

    # 최적값을 찾기 위한 기준
    es_score = 1000000000
    best_period = 0
    # 각 store_id 최적의 seasonal period를 찾기 위한 탐색
    for p in range(2,(len(train)//2) + 1):
        try:
            es_model = ExponentialSmoothing(np.array(train.RUN), seasonal_periods=p, trend='add', seasonal='add')
            es_result = es_model.fit()
            es_pred = es_result.forecast(len(test))

            if es_score > mae(test.RUN, es_pred):
                es_score = mae(test.RUN, es_pred)
                best_period = p
        except:
            continue
    print("Mean value of store_id 0 :", df_set.RUN.mean())
    print("MAE Score of test :", es_score)
    print("Best seasonal period :", best_period)

    plot_train_test_pred_graph(train, test, es_pred)
    
    return es_pred
```

```python
best_pred_list=[]
for team in team_list:
    print(team)
    pred=HoltsWintersExpSmoothing(hiter_df, team)
    best_pred_list.append(pred)
```

## ARMA

```python
def myARMA(df,team):
    team = df[df['T_ID']==team]
    df_set = team

    size = int(len(df_set) * 0.85)
    train = df_set[:size]
    test = df_set[size:]

    set_arma = sm.tsa.ARMA(train.RUN, (0, 2))
    set_result = set_arma.fit()

    set_output = set_result.forecast(len(test))
    set_pred = set_output[0]
    
    print("Mean value of store_id 0 :", df_set.RUN.mean())
    print("MAE Score of test :", mae(test.RUN, set_pred))

    plot_train_test_pred_graph(train, test, set_pred)
    
    return set_pred
```

```python
pred_list=[]
for team in team_list:
    print(team)
    pred=myARMA(hiter_df, team)
    pred_list.append(pred)
```
