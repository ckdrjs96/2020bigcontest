## 방어율 예측

#### 1) 각 경기별 ER, INN2을 각각 예측해서 ERA을 구하는 방법 2) 각 경기별 ERA를 구해서 평균을 내는 방법 이 있는데 INN2는 대부분 27이라 예측에서 오류가 발생할 확률이 높을 것 같고 ER은 실험해보니 ERA랑 그 prediction 오차가 크게 차이가 나지 않아서 2번 방법을 사용하고자 함

### EDA

```python
from sklearn.preprocessing import LabelEncoder

def team_pitcher_eda(year, team_name):
    year_index = year - 2016 
    
    # team_pitcher data에 해당 경기의 해당 팀의 홈 어웨이 여부 column 추가
    data_games_year = data_games[year_index]
    data_games_year = data_games_year[['G_ID', 'VISIT_KEY', 'HOME_KEY']]
    data_games_year = data_games_year.set_index('G_ID')
    
    data_team_pitcher_year = data_team_pitcher[year_index]
    data_team_pitcher_year = pd.merge(data_team_pitcher_year, data_games_year, how='left', on=['G_ID'])

    data_team_pitcher_year_team = data_team_pitcher_year[data_team_pitcher_year.T_ID == team_name]
    df = data_team_pitcher_year_team
    df['HOME_KEY'] = df['HOME_KEY'].map(lambda x: 1 if x == team_name else 0 )
    df = df.drop(columns = ['VISIT_KEY', 'GDAY_DS'])
    
    # LabelEncoding
    encoder = LabelEncoder()

    df['VS_T_ID'] = encoder.fit_transform(df['VS_T_ID']) #상대편
    df['TB_SC'] = encoder.fit_transform(df['TB_SC']) #이닝 초/말
    df['WLS'] = encoder.fit_transform(df['WLS']) #승패무 여부

    # 시계열 데이터(경기당 방어율 계산)이기 때문에 경기 코드를 index로
    df = df.set_index('G_ID')

    # 경기당 방어율 column 생성
    df['ERA'] = df['ER'] * 27 / df['INN2']

    df = df.drop(columns=['T_ID'])
    df_year_team_name = df
    
    return df_year_team_name
```

```python
df_2016_LG = team_pitcher_eda(2016,'LG')
df_2016_LG
```

### Util

```python
import itertools
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.api import SimpleExpSmoothing, Holt, ExponentialSmoothing

import warnings
warnings.filterwarnings(action='ignore')
```

```python
from sklearn.metrics import make_scorer

def mae(prediction, correct):
    prediction = np.array(prediction)
    correct = np.array(correct)
    
    difference = correct - prediction
    abs_val = abs(difference)
    
    score = abs_val.mean()
    
    return score

mae_scorer = make_scorer(mae)
mae_scorer
```

```python
def mse(prediction, correct):
    prediction = np.array(prediction)
    correct = np.array(correct)
    
    difference = correct - prediction
    sqr_val = difference ** 2
    
    score = sqr_val.mean()
    
    return score

mse_scorer = make_scorer(mse)
mse_scorer
```

```python
def plot_train_test_pred_graph(trainset, testset, pred):
    plt.figure(figsize=(15,3))
    plt.plot(trainset.ERA, label='train')
    plt.plot(testset.ERA, label='test')
    plt.plot(testset.index, pred, label='prediction')
    plt.legend()
    plt.show()
```

### 모델링

###  Holt-Winter's Exponential Smoothing

```python
class HoltWinter:
    def __init__(self, df):
        super(HoltWinter, self).__init__()
        
        self.df_set = df
        self.train = self.df_set.iloc[:-20]
        self.test = self.df_set.iloc[-20:]
        
    def run(self):
        es_score = 1000000000
        best_period = 0
        # 각 ERA 최적의 seasonal period를 찾기 위한 탐색
        for p in range(2,(len(train)//2) + 1):
            try:
                es_model = ExponentialSmoothing(np.array(self.train.ERA), seasonal_periods=p, trend='add', seasonal='add')
                es_result = es_model.fit()
                es_pred = es_result.forecast(len(self.test))

                if es_score > mae(self.test.ERA, es_pred):
                    es_score = mae(self.test.ERA, es_pred)
                    best_period = p
            except:
                continue
                
        return self.test.ERA.mean(), es_pred.mean()
        
    def visualize(self):
        
        self.run()
        print("Mean value of ERA 0 :", self.df_set.ERA.mean())
        print("MAE Score of test :", es_score)
        print("Best seasonal period :", best_period)
        print("Pred ERA mean :", es_pred.mean())
        print("Test ERA mean :", self.test.ERA.mean())
        
        plot_train_test_pred_graph(train, test, es_pred)
            
        return self.test.ERA.mean(), es_pred.mean()
                   
        
```

```python
holtwinter_2016_LG = HoltWinter(df_2016_LG)
holtwinter_2016_LG.visualize()
```

### ARIMA

```python
class Arima:
    def __init__(self, df):
        super(Arima, self).__init__()
        
        self.df_set = df
        self.train = self.df_set.iloc[:-20]
        self.test = self.df_set.iloc[-20:]
    
    def run(self):
        p = list(range(0, 6))
        d = [0, 1, 2]
        q = list(range(0, 6))

        pdq = list(itertools.product(p, d, q))
        best_score = 10000000
        best_param = 0
        for param in pdq:
            try:
                arima_model = ARIMA(self.train.ERA.values, order=param)
                result = arima_model.fit()
                if result.aic < best_score:
                    best_score = result.aic
                    best_param = param
            except:
                continue
                
        set_arima = ARIMA(self.df_set.ERA.values, order=best_param)
        set_result = set_arima.fit()
        set_pred = set_result.forecast(len(self.test))[0]
        
        return self.test.ERA.mean(), set_pred.mean()
    
    def visualize(self):
        
        self.run()
        print("Mean value of ERA :", self.df_set.ERA.mean())
        print("AIC Score of test :", best_score)
        print("Best parameter of (p, d, q): ",best_param)
        print("Pred ERA mean :", set_pred.mean())
        print("Test ERA mean :", self.test.ERA.mean())

        plot_train_test_pred_graph(train, test, set_pred)
        
        return self.test.ERA.mean(), set_pred.mean()
```

```python
arima_2016_LG = Arima(df_2016_LG)
arima_2016_LG.visualize()
```

#### 위 EDA 방법으로 각 연도별, 팀별 마지막 20경기에 대한 실제 평균 ERA값(Test.ERA)과 모델(HoltWinter,ARIMA)이 예측한 prediction 평균 ERA값을 차트로 비교해보았다

```python
dict = {
    'year': [],
    'team': [],
    'Test.ERA': [],
    'Pred.ERA.HOLT' : [],
    'Pred.ERA.ARIMA' : []
}

#2016~2019, OB, LG, WO, SK, KT, HH, HT, SS, LT, NC
for i in range(4):
    year = 2016+i
    for team in ['OB', 'LG', 'WO', 'SK', 'KT', 'HH', 'HT', 'SS', 'LT', 'NC']:
        dict['year'].append(year)
        dict['team'].append(team)
        
        df = team_pitcher_eda(year,team)
        holtwinter_model = HoltWinter(df)
        holtwinter_test, holtwinter_pred = holtwinter_model.run()
        
        arima_model = Arima(df)
        arima_test, arima_pred = arima_model.run()
        
        dict['Test.ERA'].append(holtwinter_test)
        dict['Pred.ERA.HOLT'].append(holtwinter_pred)
        dict['Pred.ERA.ARIMA'].append(arima_pred)
        
df = pd.DataFrame(dict)
df
```

```python
#모델별 오차의 절대값 비교 
df['holt_abs_difference'] = abs(df['Test.ERA']-df['Pred.ERA.HOLT'])
df['arima_abs_difference'] = abs(df['Test.ERA']-df['Pred.ERA.ARIMA'])
```

```python
print(df['holt_abs_difference'].mean())
print(df['arima_abs_difference'].mean())
```
